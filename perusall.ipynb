```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('/content/sample_data/EvaluaciondelaHerramienta.csv', encoding='latin-1')


data.rename(columns={
    'Fecha': 'Fecha',
    '1.Nombre de la Estudiante': 'Nombre',
    'Género': 'Genero',
    '2.Edad': 'Edad',
    '3. Correo Electrónico': 'Correo',
    '4. Nivel de la carrera': 'NivelCarrera',
    '5. ¿Usted trabaja?': 'Trabaja',
    '6. ¿Ya conocía la herramienta Perussall?': 'ConocimientoPerussall',
    '7.¿Considera la herramienta útil?': 'Utilidad',
    '8. ¿Por qué?': 'RazonUtilidad',
    '9. ¿Considera que esta herramienta favorece en su proceso de enseñanza y aprendizaje?': 'FavorcAprendizaje',
    '10. ¿Recomendaría el uso de esta herramienta en futuras ediciones del curso? ¿Porqué?': 'RecomFutEdiciones',
    '11. Marque la mejor opción que corresponde para usted cada una de las siguientes consideraciones: [¿Perusall hizo que fuera más probable que hicieras la lectura en lugar de leerla por tu cuenta?]': 'ProbabilidadLectura',
    '11. Marque la mejor opción que corresponde para usted cada una de las siguientes consideraciones: [¿Te ayudó Perusall a ENTENDER la lectura mejor que leerla por tu cuenta?]': 'AyudaComprension',
    '11. Marque la mejor opción que corresponde para usted cada una de las siguientes consideraciones: [¿La herramienta lo motiva a seguir utilizando esta herramienta?]': 'MotivacionUso',
    '11. Marque la mejor opción que corresponde para usted cada una de las siguientes consideraciones: [¿Considera que esta herramienta es fácil de usar?]': 'FacilidaddeUso',
    '11. Marque la mejor opción que corresponde para usted cada una de las siguientes consideraciones: [¿Recomienda hacer uso de esta herramienta en otros cursos?]': 'RecomUsoOtCursos',
    '12. Número de Comentarios (aportaciones realizadas en la herramienta)': 'NumeroComentarios',
    '13. ¿Se siente mejor haciendo uso de Perussal para realizar sus comentarios o prefiere hacer los comentarios de forma verbal?': 'PrefComentarios',
    '14. Realice un comentario sobre el uso de esta herramienta': 'Comentario'
}, inplace=True)


print(data.columns)

```

```python
import matplotlib.pyplot as plt

# Contar el número de respuestas en cada categoría
utilidad_counts = data['Utilidad'].value_counts()

# Calcular los porcentajes
total_responses = utilidad_counts.sum()
utilidad_percentages = utilidad_counts / total_responses * 100

# Crear un gráfico de barras
plt.bar(utilidad_counts.index, utilidad_counts.values)

# Agregar los porcentajes a las barras
for i, value in enumerate(utilidad_counts.values):
    plt.text(i, value + 1, f'{utilidad_percentages[i]:.1f}%', ha='center')

# Personalizar el gráfico
plt.xlabel('Utilidad')
plt.ylabel('Cantidad de respuestas')
plt.title('¿Considera que esta herramienta es útil?')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras de la distribución de edades
age_counts = data['Edad'].value_counts().sort_values(ascending=False)
# Crear una figura y ejes más grandes
plt.figure(figsize=(10, 6))

# Crear el gráfico de barras
age_counts.plot(kind='bar', color='skyblue')

# Personalizar el gráfico
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edades')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
gender_counts = data['Genero'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
plt.axis('equal')
plt.ylabel('')
plt.title('Distribución de Estudiantes por Género')
plt.show()
```

```python
# Gráfico de pastel de la proporción de estudiantes que consideran la herramienta útil
utilidad_counts = data['Utilidad'].value_counts()
labels = utilidad_counts.index
sizes = utilidad_counts.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Proporción de Estudiantes que consideran la herramienta Útil')
plt.show()
```

```python
utility_by_career = data.groupby('NivelCarrera')['Utilidad'].value_counts().unstack()
utility_by_career.plot(kind='bar', stacked=True)
plt.xlabel('Nivel de la Carrera')
plt.ylabel('Cantidad de Estudiantes')
plt.title('Utilidad de Perussall según el Nivel de la Carrera')
plt.legend(title='Utilidad')
plt.show()
```

```python
#Gráfico de pastel: Proporción de estudiantes que trabajan
work_counts = data['Trabaja'].value_counts()
work_counts.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Proporción de Estudiantes que Trabajan')
plt.show()
```
