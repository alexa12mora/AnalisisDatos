```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/sample_data/CSVInvestigation (1).csv')
data = pd.DataFrame(df)
data = data.astype({"Value": float})
dataFiltered2018 = data.loc[(data['Year']==2018)]

#print ("5 paises con mayor indice criminal")
print()
print()
dataFiltered=data.loc[(data['Series'] == 'Intentional homicide rates per 100,000')&(data['Year']==2018)]
print("A-Los 5 paises con mayor indice de homicidios intencionales en el año 2018 son: ")
print(dataFiltered.sort_values(by=['Value'],ascending = False)[['Country','Value']].head(5))


#print ("5 paises con mayor indice criminal")
print()
print()
print("B-Los 5 paises con menor indice de homicidios intencionales en el año 2018 son: ")
print(dataFiltered.sort_values(by=['Value'],ascending = True)[['Country','Value']].head(5))

print()
print()
print("C- Crimen mas repetido por pais en el 2018")
print(dataFiltered2018['Series'].value_counts())

print()
print()
print("D- Crimen con el porcentaje mas alto en Costa Rica en el 2005")
dataFilteredCR=data.loc[(data['Country'] == 'Costa Rica')&(data['Year']==2005)]
print(dataFilteredCR.sort_values(by=['Value'],ascending = False)[['Country','Series','Value']])

print()
print()
print("E- En el año 2010, cual es la tasa de homicidios en Costa Rica, Panama y Nicaragua")
print()
print()

registro_general = df.loc[(((df['Country']=='Costa Rica') | (df['Country']=='Panama') 
| (df['Country']=='Nicaragua'))
& (df['Series'] == 'Intentional homicide rates per 100,000')
& (df['Year']==2010))][['Country','Value']]
print(registro_general)

#porcentajes de homicidio intencional por CR por año
print()
print()
print('F-Porcentajes de homicidio intencional por CR por año')
dataFilteredCRHomicides=data.loc[(data['Country'] =="Costa Rica")&
                                 (data['Series'] =="Intentional homicide rates per 100,000")]
print(dataFilteredCRHomicides.sort_values(by=['Year'], ascending=True)[['Country','Year','Value']])

#obtener el anio mas conflictivo en general
print()
print()
print("G.Obtener el año más conflictivo en general.")
registro_annos = df["Year"]
anno_conflictivo = registro_annos.mode()
print (anno_conflictivo)

#GENERO MAS AFECTADO EN CRIMENES EN PROMEDIO EN EL ANO 2018
print()
print()
print('H-Genero mas afectado en crimenes en promedio en el año 2018')
dataFilteredByGender=dataFiltered2018.loc[((
    dataFiltered2018['Series'] =="Percentage of male and female intentional homicide victims, Male")
    |(dataFiltered2018['Series']=="Percentage of male and female intentional homicide victims, Female"))
    ].groupby('Series')['Value']
print(dataFilteredByGender.mean())


print("Grafico Pastel-Porcentaje de homicidios por genero")

dataFilteredByGender.mean().plot.pie(y='Serie',figsize = (8,8),autopct='%1.0f%%')
plt.show()


print("Grafico de histograma de la taza de homicidios de CR,Panama y Nicaragua")

registro_general = data.loc[(((df['Country']=='Costa Rica') | (df['Country']=='Panama') 
| (df['Country']=='Nicaragua'))
& (df['Series'] == 'Intentional homicide rates per 100,000')
& (df['Year']==2018))][['Country','Value']]

registro_general.plot.barh(x="Country", y="Value")
```
