```python
import folium 
from folium.plugins import MiniMap
from branca.element import Figure
Casa = folium.Map(location=(8.888317, -82.774148))
Casa

popuptext = '<b>Casa</b>'
my_map = folium.Map(location=[8.888317, -82.774148], zoom_start=16,tiles='stamenterrain')

folium.Marker(location=[8.888317, -82.774148],popup=popuptext,icon=folium.Icon(color="orange",icon="ok-sing")).add_to(my_map)
folium.Circle(location=[8.888317, -82.774148],color="purple",weigth=6,fill_color="red",fill_opacity=0.5,tooltip="Heredia").add_to(my_map)

minimap = MiniMap(tile_layer='stamenterrain',position = 'bottomrigth')
my_map.add_child(minimap)
my_map.save('C:folium.html')
my_map

fig = Figure(width=1000,height=500)
fig.add_child(my_map)
fig
```
